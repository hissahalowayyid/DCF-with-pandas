```python
import pandas as pd
```

```python
Revenue = 222 
Revenue Growth Factor = 18% of prior period Revenue
EBITDA Margin = 45.0% of current Revenue
Taxes = 21% of current EDBITDA
Capital Expenditure = 3% of current Revenue
Working Capital Increase = 6% of Revenue
EBITDA Exit Multiple = 32.70
Discount Factor = 0.94 (6%) 
```

```python
years = ['2021','2022', '2023', '2024', '2025', '2026', '2027', '2028', '2029','2030','2031','2032']
df = pd.DataFrame(index = years, columns=['Revenue','EBITDA','Taxes','CAPX','NWC','Cash_Flow','Discount_Factor','PV_FCF'])
```

```python
#Revenue
df['Revenue'][0] = 222
growth_rate = 0.18
for year in range(1, 12):    
    df['Revenue'][year] = df['Revenue'][year-1] * (1 + growth_rate)  
df
```

```python
#EBITDA
ebitda_margin = 0.45
for year in range(0, 12):    
    df['EBITDA'][year] = df['Revenue'][year] * ebitda_margin  
df
```

```python
#Tax
tax_rate = 0.21
#tax_payment = -ebitda * tax_rate
for year in range(0, 12):    
    df['Taxes'][year] = -df['EBITDA'][year] * tax_rate  
df
```

```python
#Capital Expenditures
capex_percent =.03 
#capex = -(Revenue * capex_percent)
for year in range(0, 12):    
    df['CAPX'][year] = -(df['Revenue'][year] *capex_percent )  
df
```

```python
#Net Working Capital
nwc_percent = 0.06  #Working Capital Increase: 6% of Revenue
#nwc = Revenue * nwc_percent
for year in range(0, 12):    
    df['NWC'][year] = (-df['Revenue'][year] *nwc_percent )  
df
```

```python
#Discount Factor

df['Discount_Factor'] = [(0.94 ** i) for i in range(len(df))]

df
```

```python
#Cash flow 
#Cash_Flow=ebit + tax + capex + change_in_nwc

for year in range(0, 12):    
    df['Cash_Flow'][year] = (df['EBITDA'][year] + df['Taxes'][year] + df['CAPX'][year]+ df['NWC'][year])  
df
```

```python
#Present value = discount factor * Cash flows
for year in range(0, 12):    
    df['PV_FCF'][year] = (df['Discount_Factor'][year] * df['Cash_Flow'][year])  
df
```

```python
#Terminal value = EBITDA Exit Multiple * df['EBITDA'].iloc[2031]
EDBITA_exit_multiple= 32.70
TerminalValue= EDBITA_exit_multiple * df['EBITDA'][-1]
TerminalValue
```

```python
#Sum of PV_FCF
Sum_of_PV_FCF=df.PV_FCF[1:-1].sum()
Sum_of_PV_FCF
```

```python
PV_terminal_value = TerminalValue * df.Discount_Factor[-2]
PV_terminal_value
```

```python
present_value_of_asset = Sum_of_PV_FCF + PV_terminal_value
present_value_of_asset
```

```python
statement = f"""
Our projected Present Value of Projected Future Cash Flows are: {Sum_of_PV_FCF:.2f}.

We are providing revenue guidance in 2032 to be: {df.loc["2032", "Revenue"].values[0]:.2f}

With an estimated EBITDA around: {df.loc["2032", "EBITDA"][0]:.2f}.

The terminal value a this time given a multiple of {EDBITA_exit_multiple:.2f} and the above EBITDA for 2030, would equal: {PV_terminal_value:.2f}.

"""

print(statement)
```
